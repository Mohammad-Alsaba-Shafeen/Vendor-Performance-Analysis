### Creating Summary

    - purchase transactions made by vendors.
    - sales transaction data
    - freight costs for each vendor
    - actual product prices from vendors

```python
# Step 1: Perform the join (you've already done this)
merged_df = pd.merge(purchases, purchase_prices, on='Brand', how='inner')
merged_df.columns
```

```python
# Step 2: Group and aggregate
summary = merged_df.groupby(
    ['VendorNumber_x', 'VendorName_x', 'Brand', 'PurchasePrice_x', 'Volume', 'Price'],
    as_index=False
).agg(
    TotalPurchaseQuantity=('Quantity', 'sum'),
    TotalPurchaseDollars=('Dollars', 'sum')
)

# Step 3: Rename 'Price' to 'ActualPrice' for clarity
summary = summary.rename(columns={'VendorNumber_x': 'VendorNumber',
                                  'VendorName_x': 'VendorName',
                                  'PurchasePrice_x': 'PurchasePrice',
                                  'Price': 'ActualPrice'})

# Step 4: Sort by TotalPurchaseDollars
summary = summary.sort_values(by='TotalPurchaseDollars')
```

```python
summary
```

```python
purchase_summary = summary[summary['PurchasePrice'] > 0]
purchase_summary
```

```python
sales.columns
```

```python
sales_summary = sales.groupby(['VendorNo', 'Brand'], as_index=False).agg(
    TotalSalesDollars=('SalesDollars', 'sum'),
    TotalSalesPrice=('SalesPrice', 'sum'),
    TotalSalesQuantity=('SalesQuantity', 'sum'),
    TotalExciseTax=('ExciseTax', 'sum')
)
sales_summary=sales_summary.sort_values(by='TotalSalesDollars')
sales_summary
```

```python
import pandas as pd

# ─────────────────────────────────────────────
# 1️⃣  FREIGHT SUMMARY  (vendor_invoice ➜ freight_summary)
# ─────────────────────────────────────────────
freight_summary = (
    vendor_invoice
    .groupby('VendorNumber', as_index=False)
    .agg(FreightCost=('Freight', 'sum'))
)

# ─────────────────────────────────────────────
# 2️⃣  PURCHASE SUMMARY  (purchases + purchase_prices ➜ purchase_summary)
# ─────────────────────────────────────────────
# Inner‑join on Brand
purch_pp_merged = purchases.merge(
    purchase_prices[['Brand', 'Price', 'Volume']],  # keep only what we need from pp
    on='Brand',
    how='inner',
    suffixes=('', '_pp')         # avoid x/y clutter
)

# Filter PurchasePrice > 0
purch_pp_merged = purch_pp_merged[purch_pp_merged['PurchasePrice'] > 0]

# Aggregate
purchase_summary = (
    purch_pp_merged
    .groupby(
        [
            'VendorNumber',
            'VendorName',
            'Brand',
            'Description',
            'PurchasePrice',      # purchase price from purchases
            'Price',              # actual price from purchase_prices
            'Volume'
        ],
        as_index=False
    )
    .agg(
        TotalPurchaseQuantity=('Quantity', 'sum'),
        TotalPurchaseDollars=('Dollars',  'sum')
    )
    .rename(columns={'Price': 'ActualPrice'})
)

# ─────────────────────────────────────────────
# 3️⃣  SALES SUMMARY  (sales ➜ sales_summary)
# ─────────────────────────────────────────────
sales_summary = (
    sales
    .groupby(['VendorNo', 'Brand'], as_index=False)
    .agg(
        TotalSalesDollars=('SalesDollars', 'sum'),
        TotalSalesPrice  =('SalesPrice',   'sum'),
        TotalSalesQuantity=('SalesQuantity', 'sum'),
        TotalExciseTax   =('ExciseTax',    'sum')
    )
    .rename(columns={'VendorNo': 'VendorNumber'})
)

# ─────────────────────────────────────────────
# 4️⃣  FINAL JOIN & SORT
# ─────────────────────────────────────────────
vendor_sales_summary = (
    purchase_summary
    .merge(sales_summary, on=['VendorNumber', 'Brand'], how='left')
    .merge(freight_summary, on='VendorNumber',           how='left')
    .sort_values(by='TotalPurchaseDollars', ascending=False)
    .reset_index(drop=True)
)

# ─────────────────────────────────────────────
# 5️⃣  Peek at the result
# ─────────────────────────────────────────────
vendor_sales_summary
```

### What is the 95% confidence intervals for profit margins of top-performing and low-performing vendors

```python
top_threshold=df['TotalSalesDollars'].quantile(0.75)
low_threshold=df['TotalSalesDollars'].quantile(0.25)
```

```python
top_vendors=df[df['TotalSalesDollars']>=top_threshold]['ProfitMargin'].dropna()
low_vendors=df[df['TotalSalesDollars']<=low_threshold]['ProfitMargin'].dropna()
```

```python
low_vendors
```

```python

def confidence_interval(data,confidence=0.95):
    mean_val=np.mean(data)
    std_err=np.std(data,ddof=1)/np.sqrt(len(data))
    t_critical=stats.t.ppf((1+confidence)/2,df=len(data)-1)
    margin_of_error=t_critical*std_err
    return mean_val,mean_val-margin_of_error,mean_val+margin_of_error
```

```python
top_mean,top_lower,top_upper=confidence_interval(top_vendors)
low_mean,low_lower,low_upper=confidence_interval(low_vendors)

print(f"Top Vendors 95% CI: ({top_lower:.2f},{top_upper:.2f}),Mean:{top_mean:.2f}")
print(f"Low Vendors 95% CI: ({low_lower:.2f},{low_upper:.2f}),Mean:{low_mean:.2f}")

plt.figure(figsize=(12,6))

# Top Vendors Plot
sns.histplot(top_vendors,kde=True,color='blue',bins=30,alpha=0.5,label='Top Vendors')
plt.axvline(top_lower,color='blue',linestyle='--',label=f"Top Lower: {top_lower:.2f}")
plt.axvline(top_upper,color='blue',linestyle='--',label=f"Top Upper: {top_upper:.2f}")
plt.axvline(top_mean,color='blue',linestyle='-',label=f"Top Mean {top_mean:.2f}")

#Low Vendors plot
sns.histplot(low_vendors,kde=True,color='red',bins=30,alpha=0.5,label="Low Vendors")
plt.axvline(low_lower,color='blue',linestyle='--',label=f"Top Lower: {low_lower:.2f}")
plt.axvline(low_upper,color='blue',linestyle='--',label=f"Top Upper: {low_upper:.2f}")
plt.axvline(low_mean,color='blue',linestyle='-',label=f"Top Mean {low_mean:.2f}")

#Finalize Plot
plt.title("Confidence Interval Comparision Top vs. Low Vendors (Profit Margin)")
plt.xlabel("Profit Margin (%)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

- The confidence interval for low-performing vendors is significantly higher than that of top-performing vendors.
- This suggests that vendors with lower sales tend to maintain higher profit margins,potentially due to premium pricing or lower operational costs.
- For High-Performing Vendors: If they aim to improve profitability,they could explore selective price adjustements,cost optimization,or bunding strategies.
- For Low-Performing Vendors: Despite higher margins,their low sales volume might indicate a need for better marketing,competitive pricing, or improved.
